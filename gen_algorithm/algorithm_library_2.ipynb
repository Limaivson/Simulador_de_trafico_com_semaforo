# Imports

```python
import pygad
import numpy as np
from fitness_function import ff
from model_traffic_light import TrafficLight
```

# Preparing the fitness_func Parameter

```python
def fitness_func(ga_instance, solution, solution_idx):
    ways = 2
    light = TrafficLight(solution[0], solution[1], solution[2])
    fitness_value = 1/ff(light, ways)
    return fitness_value
```

# Preparing Other Parameters

```python
num_generations = 100
num_parents_mating = 4

fitness_function = fitness_func

sol_per_pop = 8
num_genes = 3

init_range_low = 0
init_range_high = 100

parent_selection_type = "sss"
keep_parents = 2

crossover_type = "single_point"

mutation_type = "random"
mutation_percent_genes = 30
```

# Initial solution

```python
# light_init = TrafficLight(50, 10, 0.01)

# initial_solution = [light_init.t_max, light_init.t_min, light_init.coef]
```

# Create an Instance of the pygad.GA Class

```python
ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       fitness_func=fitness_function,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       init_range_low=init_range_low,
                       init_range_high=init_range_high,
                       parent_selection_type=parent_selection_type,
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes)
```

# Run the Genetic Algorithm

```python
ga_instance.run()
```

# Plotting Results

```python
ga_instance.plot_fitness()
```

# Information about the Best Solution

```python
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))
print("Index of the best solution : {solution_idx}".format(solution_idx=solution_idx))
```

```python
if ga_instance.best_solution_generation != -1:
    print("Best fitness value reached after {best_solution_generation} generations.".format(best_solution_generation=ga_instance.best_solution_generation))
```
